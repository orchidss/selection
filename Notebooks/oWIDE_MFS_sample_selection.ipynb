```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np
from scipy.special import erf
from scipy.special import ndtri, ndtr
from tqdm import tqdm

from astropy.table import Table, vstack, Column, join
from astropy.coordinates import SkyCoord, SkyOffsetFrame, Angle
import astropy.units as u

from astropy.cosmology import FlatLambdaCDM
from mocpy import MOC

cos = FlatLambdaCDM(H0=70, Om0=0.3)

import eazy

# Symlink templates & filters from the eazy-code repository
try:
    print('EAZYCODE = '+os.getenv('EAZYCODE'))
except:
    pass

if not os.path.exists('templates'):
    eazy.symlink_eazy_inputs() 

   
# quiet numpy/astropy warnings
import warnings
from astropy.utils.exceptions import AstropyWarning

np.seterr(all='ignore')
warnings.simplefilter('ignore', category=AstropyWarning)

import smpy.smpy as S
from scipy.interpolate import griddata
import astropy.constants as c
from sklearn.neighbors import KNeighborsClassifier

from dustmaps.config import config
config.reset()
from dustmaps.sfd import SFDQuery

def calcflux(SED, wave, filt, units):
    """ Convolve synthetic SEDs with a given filter

    Arguments
    ---------
        SED : numpy.array
            Grid of synthetic spectra
        filt : '~smpy.Filter' class
            Filter through which to convolve SED grid
        z : float
            Redshift at which models are to be observed
        dl : '~astropy.units.Quantity'
            Luminosity distance corresponding to redshift(z) in given
            cosmology.
        units : '~astropy.units'
            Desired output flux units (in spectral flux density)

    Returns
    -------
        Flux : '~astropy.units.Quantity'
            Spectral flux density, with exact units as given by 'units'

    """
    # Find SED wavelength entries within filter range
    wff = np.logical_and(filt.wave[0] < wave,
                         wave < filt.wave[-1])
    wft = wave[wff]

    # Interpolate to find throughput values at new wavelength points
    tpt = griddata(filt.wave, filt.response, wft)

    # Join arrays and sort w.r.t to wf
    # Also replace units stripped by concatenate
    wf = np.array(np.concatenate((filt.wave, wft))) * u.AA
    tp = np.concatenate((filt.response, tpt))

    order = np.argsort(wf)
    wf = wf[order]
    tp = tp[order]

    # Interpolate redshifted SED and LyAbs at new wavelength points
    sed = griddata(wave, SED.T, wf).T * SED.unit
    #lyabs = griddata(wave, self.lyman_abs, wf)

    # Calculate f_nu mean
    # Integrate SED through filter, as per BC03 Fortran
    # As: f_nu=int(dnu Fnu Rnu/h*nu)/int(dnu Rnu/h*nu)
    # ie: f_nu=int(dlm Flm Rlm lm / c)/int(dlm Rlm/lm)
    top = np.trapz(sed * tp * wf /
                   c.c.to(u.AA / u.s), wf)
    bottom = np.trapz(tp / wf, wf)
    #area = (4 * np.pi * (dl ** 2))
    Flux = top / bottom #/ (1 + z) / area

    return Flux.to(units)


def pbelowz(zphot, zerr, zcut):
    dz = zphot - zcut
    sigma_abs = (np.abs(dz)/zerr)

    psigma_abs = ndtr(sigma_abs)
    
    psigma = np.copy(psigma_abs)
    psigma[dz > 0.] = 1-psigma_abs[dz > 0.]
    
    return psigma
```

## Overall outline

The broad steps in the final ORCHIDSS MFS sample selection are as follows:
1. Make photo-z selection from DESI Legacy Surveys (DR8) Photometry and associated photometric redshifts (Duncan 2022)
2. Perform template fitting at fixed `z_phot` to measure UVJ colours and SFR estimates for sample selection and template matching to the S7 WAVES template set. (Based on sSFR and dust extinction since full WISE photometry not available).
3. Make sample selection based on SF UVJ cut & $M_B$ cut with magnitude limited supplement.
4. Assign templates and construct 4MOST formatted catalogue for submission.

```python
mfs = Table.read('../../../catalogues/mfs_pointingcentres_2018_radec.txt', format='ascii.commented_header')
mfs_coord = SkyCoord(mfs['RA'], mfs['Dec'], unit='deg')

mfs_mocs = [MOC.from_cone(row.ra, row.dec, radius=30*u.arcmin, max_depth=15) for row in mfs_coord]
mfs_moc = mfs_mocs[0]
for i in range(len(mfs_mocs)-1):
    mfs_moc = mfs_moc.union(mfs_mocs[i+1])
    
#mfs_moc.write('MOCS/MFS_orchidss_wide.moc', overwrite=True)
```

```python
lsdr8 = Table.read('../InputCatalogues/optical/ls_dr8_mfs_region.fits')

lsdr8_pz = Table.read('../InputCatalogues/photoz/lsdr8_pz_mfs_region.fits')
lsdr8_pz_coord = SkyCoord(lsdr8_pz['ra'], lsdr8_pz['dec'], unit='deg')

lsdr8_morph = Table.read('../InputCatalogues/optical/ls_dr8_mfs_morph.fits')
in_mfs = mfs_moc.contains(lsdr8_pz_coord.ra, lsdr8_pz_coord.dec)
```

```python
wide_prob = pbelowz(lsdr8_pz['zphot'], lsdr8_pz['zphot_err'], 0.57)

cut = ((wide_prob > 0.80) * # High photo-z selection probability
       (lsdr8_pz['flag_qual'] == 1) * # Reliable redshift
       (lsdr8_pz['zphot'] > 0.) * # Physical redshift for any missed by above
       in_mfs) # In correct region

lsdr8_merged = join(lsdr8_pz[cut], lsdr8, 
                    keys=['release', 'brickid', 'objid', 'ra', 'dec', 'type'], 
                    join_type='left')
lsdr8_merged = join(lsdr8_merged, lsdr8_morph, 
                    keys=['release', 'brickid', 'objid', 'ra', 'dec', 'type'], 
                    join_type='left')

zmag_tot = 22.5 -2.5*np.log10((lsdr8_merged['flux_z']/lsdr8_merged['mw_transmission_z']))
rmag_tot = 22.5 -2.5*np.log10((lsdr8_merged['flux_r']/lsdr8_merged['mw_transmission_r']))
gmag_tot = 22.5 -2.5*np.log10((lsdr8_merged['flux_g']/lsdr8_merged['mw_transmission_g']))

snr_z = lsdr8_merged['flux_z']*np.sqrt(lsdr8_merged['flux_ivar_z'])
```

```python
nanomaggy_to_ujy = 10 ** ((23.9 - 22.5) / 2.5)

filters = ['g', 'r', 'z', 'w1', 'w2', 'w3', 'w4']

eazy_input = lsdr8_merged[['id', 'ra', 'dec', 'type']].copy()

for filt in filters:
    eazy_input[f'flux_{filt}'] = ((lsdr8_merged[f'flux_{filt}'] * nanomaggy_to_ujy) / 
                                      lsdr8_merged[f'mw_transmission_{filt}'])
    eazy_input[f'fluxerr_{filt}'] = (((1/np.sqrt(lsdr8_merged[f'flux_ivar_{filt}'])) * nanomaggy_to_ujy) / 
                                         lsdr8_merged[f'mw_transmission_{filt}'])
    

eazy_input['z_spec'] = lsdr8_merged['zphot']
eazy_input.write('lsdr8_pz_mfs_eazy_input.cat', format='ascii.commented_header', overwrite=True)
```

```python
filters = [col.split('_')[1] for col in eazy_input.colnames if col.startswith('flux_')]
print(filters)

template_sets = ['templates/sfhz/corr_sfhz_13.param']

pzruns = []

for i, templates in enumerate(template_sets[:]):
    params = {}
    params['CATALOG_FILE'] = 'lsdr8_pz_mfs_eazy_input.cat'

    params['MAIN_OUTPUT_FILE'] = 'bc_filters'

    # Galactic extinction
    params['MW_EBV'] = 0.00
    params['CAT_HAS_EXTCORR'] = True

    params['Z_STEP'] = 0.01
    params['Z_MIN'] = 0.0
    params['Z_MAX'] = 0.6
    params['Z_STEP_TYPE'] = 0

    params['SYS_ERR'] = 0.05
    params['APPLY_PRIOR'] = 'N'
    params['PRIOR_FILTER'] = 380
    params['PRIOR_ABZP'] = 23.9 
    params['PRIOR_FILE'] = 'templates/prior_K_TAO.dat'

    params['TEMPLATES_FILE'] = os.path.join('/', os.getenv('EAZYCODE'), templates)
    params['FIX_ZSPEC'] = True

    params['IGM_SCALE_TAU'] = 1.0

    translate_file = os.path.join(os.getenv('EAZYCODE'), 'inputs/zphot.translate')


    translate = Table()
    translate['filt'] = ['id', 'z_spec'] + [col for col in eazy_input.colnames if col.startswith('flux')][:-4]


    translate['num'] = ['id', 'z_spec',
                        'F294', 'E294',
                        'F295', 'E295',
                        'F297', 'E297',
                        'F244', 'E244',
                        'F245', 'E245',
                       ]

    translate_file = os.path.join('/', os.getenv('EAZYCODE'), 'inputs/zphot.translate')

    translate.write(translate_file, format='ascii.no_header', overwrite=True)
    
    pzfit = eazy.photoz.PhotoZ(param_file=None, translate_file=translate_file, 
                               zeropoint_file=None, 
                               params=params, 
                               load_prior=True, load_products=False)

    # Turn off error corrections derived above
    pzfit.set_sys_err(positive=True)

    # fit_parallel renamed to fit_catalog 14 May 2021
    pzfit.fit_catalog(pzfit.idx, n_proc=8)
    
    pzruns.append(pzfit)
```

```python
from astropy.cosmology import FlatLambdaCDM
from mocpy import MOC

cos = FlatLambdaCDM(H0=70, Om0=0.3)

templist = Table.read('../ETC_Templates/S7_temps_FINAL/S7_temps_FINALprops.csv')
templist = templist[templist['veldisp'] == 45]

sps_fit = pzfit.sps_parameters(cosmology=cos, extra_rf_filters=[], percentile_limits=[16, 50, 84], n_proc=8)

cat_ssfr = np.log10(sps_fit['sfr']/sps_fit['mass'])
cat_av = sps_fit['Av']

temp_ssfr = np.log10(templist['SFR']/10**(templist['logmstar']))
temp_av = templist['tau_screen'] * 1.0857
```

```python
cat_uv = -2.5*np.log10(sps_fit['restU']/sps_fit['restV'])
cat_ub = -2.5*np.log10(sps_fit['restU']/sps_fit['restB'])
cat_bv = -2.5*np.log10(sps_fit['restB']/sps_fit['restV'])

uv_col =  -2.5*np.log10(sps_fit['restU']/sps_fit['restV'])
vj_col =  -2.5*np.log10(sps_fit['restV']/sps_fit['restJ'])

passive = (uv_col > (0.88*vj_col + 0.79)) * (uv_col > 1.3) * (vj_col < 1.6)

Mb = 23.9 - 2.5*np.log10(sps_fit['restB']) - (cos.distmod(eazy_input['z_spec']).value -2.5*(np.log10(1+eazy_input['z_spec'])))
Mb_st_z = -21 + -1.03*(eazy_input['z_spec'] - 0.5)

dM = 0.7
Mb_cut = Mb < Mb_st_z +dM

mag_r = 23.9 - 2.5*np.log10(lsdr8_merged['flux_r']*nanomaggy_to_ujy)
mag_r_mw = 23.9 - 2.5*np.log10(eazy_input['flux_r'])
mag_z_mw = 23.9 - 2.5*np.log10(eazy_input['flux_z'])

optical_selection = np.logical_or(Mb_cut * np.invert(passive), mag_r_mw < 19.5) 
```

```python
mfs_area = (mfs_moc.sky_fraction*4*np.pi*u.steradian).to(u.deg**2)

print(f"Total sample: {optical_selection.sum()}")
print(f"Source density: {optical_selection.sum()/mfs_area:.0f}")

mfs_sample = lsdr8_merged[optical_selection]
```

```python
extent = np.zeros(len(mfs_sample), dtype='int')
extent_index = np.zeros(len(mfs_sample))
extent_parameter = np.zeros(len(mfs_sample))

# DeV
dev = (mfs_sample['type'] == 'DEV') + ((mfs_sample['type'] == 'COMP') * (mfs_sample['fracdev'] > 0.5))
extent[dev] = 2
extent_index[dev] = 4.0
extent_parameter[dev] = mfs_sample['shapedev_r'][dev]

# EXP
exp = ((mfs_sample['type'] == 'EXP') + (mfs_sample['type'] == 'REX') + 
       ((mfs_sample['type'] == 'COMP') * (mfs_sample['fracdev'] < 0.5)))
extent[exp] = 2
extent_index[exp] = 1.0
extent_parameter[exp] = mfs_sample['shapeexp_r'][exp]
```

### Template Matching

```python
filtset = S.LoadEAZYFilters('FILTER.RES.latest')

filters = S.FilterSet()
filters.addEAZYFilter(filtset, [152, 153, 154, 160])

templist = Table.read('../ETC_Templates/S7_temps_FINAL/S7_temps_FINALprops.csv')
templist = templist[templist['veldisp'] == 45]

temp_seds = []

for i, tempName in enumerate(tqdm(templist["tempName"][:])):
    temp = Table.read(f'../ETC_Templates/S7_temps_FINAL/{tempName}.fits')
    
    if i == 0:
        temp_lambda = temp["LAMBDA"]
        
    temp_seds.append(temp["FLUX_DENSITY"])

temp_seds = np.array(temp_seds)*temp_seds[0].unit

temp_fluxes = []
for filt in filters.filters[:-1]:
    fluxes = calcflux(temp_seds, temp_lambda, filt, units=u.uJy)
    temp_fluxes.append(fluxes)
    
temp_fluxes = np.array(temp_fluxes)*u.uJy

temp_ub = -2.5*np.log10(temp_fluxes[0]/temp_fluxes[1])
temp_bv = -2.5*np.log10(temp_fluxes[1]/temp_fluxes[2])
temp_uv = -2.5*np.log10(temp_fluxes[0]/temp_fluxes[2])
```

```python
cat_ssfr = np.log10(sps_fit['sfr']/sps_fit['mass'])
cat_av = sps_fit['Av']

temp_ssfr = np.log10(templist['SFR']/10**(templist['logmstar']))
temp_av = templist['tau_screen'] * 1.0857

Fig, Ax = plt.subplots(1, 2, figsize=(10, 4))
Ax[0].scatter(cat_ssfr, cat_ub,s=3, alpha=0.3, label='ORCHIDSS Sample')
Ax[0].scatter(temp_ssfr, temp_ub, label='S7 Templates')
Ax[0].set_xlim([-16, -7.5])
Ax[0].set_ylim([-0.2, 2.2])
Ax[0].set_xlabel(r'sSFR [Gyr$^{-1}$]', size=12)
Ax[0].set_ylabel(r'$(U-B)_{\rm{rest}}$', size=12)

Leg = Ax[0].legend(loc='upper right')

Ax[1].scatter(cat_bv, cat_ub, s=3, alpha=0.3)
Ax[1].scatter(temp_bv, temp_ub)
Ax[1].set_ylabel(r'$(U-B)_{\rm{rest}}$', size=12)
Ax[1].set_xlabel(r'$(B-V)_{\rm{rest}}$', size=12)
Ax[1].set_ylim([-0.2, 2.2])

Fig.tight_layout()
```

```python
X = np.array([temp_ssfr, temp_ub, temp_bv]).T
X[np.isinf(X)] = -100 
y = np.arange(len(temp_ssfr))

x_in = np.array([cat_ssfr, cat_ub, cat_bv]).T

x_in[x_in[:,0] < -13, 0] = -16.
for i in range(3):
    x_in[np.isinf(x_in[:,i]), i] = np.nanmedian(x_in[:,i])
    x_in[np.isnan(x_in[:,i]), i] = np.nanmedian(x_in[:,i])

knn = KNeighborsClassifier(n_neighbors=1).fit(X, y)

temp_match = knn.predict(x_in)
```

```python
name = np.array(['MFS_'+str(idn) for idn in mfs_sample['id']])

orchidsswide_mfs = Table()
orchidsswide_mfs['NAME'] = name
orchidsswide_mfs['RA'] = mfs_sample['ra']
orchidsswide_mfs['DEC'] = mfs_sample['dec']
orchidsswide_mfs['MAG'] = 22.5 - 2.5*np.log10(mfs_sample['flux_r'])
orchidsswide_mfs['MAG_ERR'] = 2.5*np.log10(np.e) * 1/(mfs_sample['flux_r']*np.sqrt(mfs_sample['flux_ivar_r']))
orchidsswide_mfs['MAG_TYPE'] = 'DECam_r_AB'
orchidsswide_mfs['REDSHIFT_ESTIMATE'] = mfs_sample['zphot']
orchidsswide_mfs['REDSHIFT_ERROR'] = mfs_sample['zphot_err']

orchidsswide_mfs['EXTENT_FLAG'] = extent
orchidsswide_mfs['EXTENT_INDEX'] = extent_index
orchidsswide_mfs['EXTENT_PARAMETER'] = extent_parameter

orchidsswide_mfs['flux_g'] = mfs_sample['flux_g']*3.631
orchidsswide_mfs['fluxerr_g'] = 1/np.sqrt(mfs_sample['flux_ivar_g'])*3.631

orchidsswide_mfs['flux_r'] = mfs_sample['flux_r']*3.631
orchidsswide_mfs['fluxerr_r'] = 1/np.sqrt(mfs_sample['flux_ivar_r'])*3.631

orchidsswide_mfs['flux_z'] = mfs_sample['flux_z']*3.631
orchidsswide_mfs['fluxerr_z'] = 1/np.sqrt(mfs_sample['flux_ivar_z'])*3.631

orchidsswide_mfs['flux_w1'] = mfs_sample['flux_w1']*3.631
orchidsswide_mfs['fluxerr_w1'] = 1/np.sqrt(mfs_sample['flux_ivar_w1'])*3.631

orchidsswide_mfs['flux_w2'] = mfs_sample['flux_w2']*3.631
orchidsswide_mfs['fluxerr_w2'] = 1/np.sqrt(mfs_sample['flux_ivar_w2'])*3.631

### Add EBV
scoord = SkyCoord(orchidsswide_mfs['RA']*u.deg, orchidsswide_mfs['DEC']*u.deg)

sfd = SFDQuery()
ebv = sfd(scoord)

orchidsswide_mfs['REDDENING'] = ebv

orchidsswide_mfs['TEMPLATE'] = templist['tempName'][temp_match[optical_selection]]
orchidsswide_mfs['FIELD'] = 'MFS'

orchidsswide_mfs.write('../TargetCatalogues/ORCHIDSSwide-mfs_20240527.fits', 
                       format='fits', overwrite=True)

sps_fit = sps_fit[optical_selection]
sps_fit['NAME'] = name
sps_fit['id'] = mfs_sample['id']
sps_fit['zphot'] = mfs_sample['zphot']
sps_fit['zphot_err'] = mfs_sample['zphot_err']

sps_fit.meta['TEMPLATES_FILE'] = ''

sps_fit.write('../TargetCatalogues/ORCHIDSSwide-mfs_20240527_sps_outputs.fits', 
              format='fits', overwrite=True)
```
